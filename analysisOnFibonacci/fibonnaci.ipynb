```python
#Fast Doubling Algorithm
def fib(n):
    if n == 0:
        return (0, 1)
    else:
        a, b = fib(n // 2)
        c = a * (b * 2 - a)
        d = a*a + b*b
        if n % 2 == 0:
            return (c, d)
        else:
            return (d, c + d)
```

```python
fast = {}
for i in range(1,10001):
    start_time = time.clock()
    final= fib(i)
    fast_time = time.clock() - start_time
    fast[i] = fast_time
```

```python
#DP
slow = {}
for j in range(1,10001):
    a = [0,1]
    start_time = time.clock()
    for i in range(2,2+j):
        x  = a[i-1]+a[i-2]
        a.append(x)
    time_is = time.clock() - start_time
    #print(a)
    slow[j] =  time_is
```
